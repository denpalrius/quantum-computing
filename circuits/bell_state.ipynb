# Bell States Circuits

```python
from qiskit import QuantumCircuit
```

 ## ‚à£ùúô+‚ü© Bell state

```python
# Create a quantum circuit with two qubits
phi_plus = QuantumCircuit(2)

# Apply Hadamard gate to the first qubit
phi_plus.h(0)

# Apply CNOT gate with the first qubit as control and the second qubit as target
phi_plus.cx(1, 0)

# Draw the circuit
phi_plus.draw('mpl')
```

 # ‚à£ùúô‚àí‚ü© Bell state

```python
# Create a quantum circuit for the phi_minus state
phi_minus = QuantumCircuit(2)
phi_minus.h(0)
phi_minus.cx(1, 0)
phi_minus.x(1)

phi_minus.draw('mpl')
```

 # ‚à£œà+‚ü© Bell state

```python
# Create a quantum circuit for the psi_plus state
psi_plus = QuantumCircuit(2)
psi_plus.h(0)
psi_plus.cx(0, 1)

psi_plus.draw('mpl')
```

 # ‚à£œà‚àí‚ü© Bell state

```python
# Create a quantum circuit for the psi_minus state
psi_minus = QuantumCircuit(2)
psi_minus.h(0)
psi_minus.cx(0, 1)
psi_minus.x(1)

psi_minus.draw('mpl')
```
